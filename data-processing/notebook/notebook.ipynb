```python
import pandas as pd
import numpy as numpy
from collections import defaultdict

game_data_set = '../dataset/steam_game_data/steam_games2023.csv'
m_cols = ['appid', 'name', 'genres', 'categories', 'recommended', 'minimum', 'price', 'short_description', 'background']

steam_game = pd.read_csv(game_data_set, engine='python')
steam_game = steam_game[m_cols]
steam_game = steam_game.fillna('')

# steam_game.genres = steam_game.genres.fillna('')
steam_game['categories'] = steam_game.categories.astype(str)
steam_game['genres'] = steam_game.genres.apply(lambda x : x.split(";"))
# steam_game.genres = steam_game.genres.fillna('')
# steam_game.genres = steam_game.genres.fillna('')

steam_game['categories'] = steam_game.categories.astype(str)
steam_game['categories'] = steam_game.categories.apply(lambda x : x.split(";"))

# check genres

steam_game

```

```python
def extract_unique_genres(df):
    unique_genres = set()  # 중복을 허용하지 않는 고유한 장르를 저장할 집합
    
    # 각 행의 장르들을 추출하여 unique_genres에 추가
    for genres_list in df['genres']:
        
        # remove empty string
        for genre in genres_list:
            if genre.strip():
                unique_genres.update(genres_list)
    
    
    return unique_genres

# 위 함수를 사용하여 데이터프레임의 장르 종류를 추출
unique_genres = extract_unique_genres(steam_game)
print(len(unique_genres))

unique_genres_df = pd.DataFrame({'genres': list(unique_genres)})
unique_genres_df.to_csv('../dataset/genres.csv')
# 추출된 장르 종류 출력
```
